```python
df['Hour_sin'] = np.sin(2 * np.pi * df['Hours'] / 24)
df['Hour_cos'] = np.cos(2 * np.pi * df['Hours'] / 24)
```

```python
X = df[['Temperature', 'Humidity', 'SquareFootage', 'Occupancy',
        'RenewableEnergy', 'HVACUsage', 'LightingUsage','Hour_cos']]
Y=df['EnergyConsumption']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
#splitting of the data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
# Check 1 - Reset index
df = df.reset_index(drop=True)

# Check 2 - Verify shapes and index alignment
print("X shape:", X.shape)
print("Y shape:", Y.shape)
print("Index aligned:", X.index.equals(Y.index))

# Check 3 - Check for NaN values
print("\nNaN in X:\n", X.isnull().sum())
print("\nNaN in Y:", Y.isnull().sum())

# Check 4 - Verify train test split shapes
print("\nX_train:", X_train.shape)
print("X_test: ", X_test.shape)
print("y_train:", Y_train.shape)
print("y_test: ", Y_test.shape)
```

```python
#creating the pipeline
pipeline=Pipeline([('scaler',StandardScaler()),
                  ('model',LinearRegression())])
```

```python
#doing cross validation
cv_scores=cross_val_score(pipeline,X_train,Y_train,cv=5,scoring='r2')
```

```python
#the results for cross validation
print(f"The cv_scores for this model is{cv_scores}")
print(f"The variance for this model is {cv_scores.var()}")
```

```python
#the model
pipeline.fit(X_train,Y_train)
```

```python
#predicting the values
Y_pred=pipeline.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
r2=r2_score(Y_pred,Y_test)
mse=mean_squared_error(Y_pred,Y_test)
rmse=root_mean_squared_error(Y_pred,Y_test)
mae=mean_absolute_error(Y_pred,Y_test)
print(r2)
print(mse)
print(rmse)
print(mae)
```

```python
#Polynomial_Features
from sklearn.preprocessing import PolynomialFeatures


poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(df[['Temperature', 'Humidity']])
```

```python
X_combined = np.hstack([X_poly, df[['SquareFootage', 'Occupancy', 
                                     'RenewableEnergy', 'HVACUsage', 
                                     'LightingUsage', 'Hour_sin', 
                                     'Hour_cos']].values])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, Y, test_size=0.2, random_state=42)
```

```python
# Step 4 — Run XGBoost
pipeline_xgb = Pipeline([
    ('scaler', StandardScaler()),
    ('model', XGBRegressor(n_estimators=300,
                           learning_rate=0.05,
                           max_depth=6,
                           random_state=42))
])
```

```python
pipeline_xgb.fit(X_train, y_train)
y_pred = pipeline_xgb.predict(X_test)
print(f"R2: {r2_score(y_test, y_pred):.4f}")
print(f"MAE:{mean_absolute_error(y_test,y_pred):.4f}")
print(f"MSE:{mean_squared_error(y_test,y_pred):.4f}")
```
